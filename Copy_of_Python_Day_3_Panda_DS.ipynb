```python
df = pd.read_excel("student.xlsx")
```

```python
df.head()
```

```python
df.info()
```

```python
# Counts missing values per column.
df.isnull().sum()
```

```python
# total missing values in df.
df.isnull().sum().sum()
```

```python
# name and class
df[["name", "class"]]
# inner [] is list of the column names. ["name", "class"]
# outer [] is for slicing.
```

```python
df.loc[11:14, ["name", "class"]]
```

```python
df.iloc[11:15, 0:2]
# selection of row and column index
# this code will return row indexes 11,12,13 and 14. and also column indexes 0 and 1.
```

```python
# Return class not Five
df[~(df["class"]=="Five")]
# first df[] is used for slicing
# second df[] is criteria
```

```python
# Return class not Five
df[(df["class"]!="Five")]
# first df[] is used for slicing
# second df[] is criteria
```

```python
# Return student with mark >=90
# df[df["mark"]>=90]
df.loc[df["mark"]>=90, :]
# : means return entire range
```

```python
# Return female student at class Four
# df[(df["gender"]=="female") & (df["class"]=="Four")]
df.loc[(df["gender"]=="female") & (df["class"]=="Four"), :]
```

```python
df.filter(items=['name', 'mark'])
```

```python
df.filter(like='ma')
#filter colum names
```

```python
df.query("mark > 60")
```

```python
df.query('gender == "male"')
```

```python
# Sorting by names
df.loc[:,:].sort_values(by="name")

# Sorting by names
df.loc[:,:].sort_values(by="mark")

# Sorting by names
df.loc[:,:].sort_values(by="name")

# Sorting by mark
df.loc[:,:].sort_values(by="mark")

# Sorting by mark in desc
df.loc[:,:].sort_values(by="mark", ascending=False).head()
```

```python
df[["name", "class"]].sort_values(by="class").head()
```

```python
# return student who are female and at class seven
df[(df["gender"]=="female") & (df["class"]=="Seven")].sort_values(by="mark")
```

```python
# return student who are female and at class seven
df[(df["gender"]=="female") & (df["class"]=="Seven")].sort_values(by="mark", ascending=False)
```

```python
df.sort_values(by = ["mark", "name"], ascending=[False,True]).head()

## How many male and female students are there?
df["gender"].value_counts()

## Class and number of student
df["class"].value_counts()
```
